```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일에서 데이터 불러오기 (파일 경로 및 파일 이름은 실제 파일에 맞게 수정해주세요)
file_path = 'date_info.csv'
df = pd.read_csv(file_path)

# 'Date' 칼럼을 날짜/시간 형식으로 변환
df['Date'] = pd.to_datetime(df['Date'])

# 'Date' 칼럼을 기준으로 연도 추출하여 새로운 칼럼 추가
df['연도'] = df['Date'].dt.year

# 연도별 데이터 수 계산
data_count = df['연도'].value_counts().sort_index()

# 그래프 그리기
plt.figure(figsize=(6, 6))
data_count.plot(kind='bar', color='lightgreen')
plt.title('Number of data by year')
plt.xlabel('Year')
plt.ylabel('Number of data')
plt.xticks(rotation=45)
plt.tight_layout()
for i, v in enumerate(data_count):
    plt.text(i, v + 50, str(v), ha='center', va='bottom', fontsize=9)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSV 파일에서 데이터 불러오기 (파일 경로 및 파일 이름은 실제 파일에 맞게 수정해주세요)
file_path = 'date_info.csv'
df = pd.read_csv(file_path)

# CS 칼럼의 값을 정렬하기 위한 순서 지정
order = ['A0', 'A1', 'A2', 'A3', 'B0', 'B1', 'B2', 'B3', 'C0', 'C1', 'C2', 'C3', 'D0', 'D1', 'D2', 'D3']

# CS 칼럼을 순서에 맞게 정렬
df['CS'] = pd.Categorical(df['CS'], categories=order, ordered=True)
df = df.sort_values(by='CS')

# 막대 그래프 그리기
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='CS', order=order, palette='Set2')
plt.title('Distribution of Current Situation')
plt.xlabel('CS Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일에서 데이터 불러오기 (파일 경로 및 파일 이름은 실제 파일에 맞게 수정해주세요)
file_path = 'arrival_info.csv'
df = pd.read_csv(file_path)

# 도착지(ARI) 코드와 실제 지역 이름을 매핑한 딕셔너리
ari_mapping = {
    2: '서울',
    31: '경기',
    32: '인천',
    33: '강원',
    41: '충남',
    42: '대전',
    43: '충북',
    44: '세종',
    51: '부산',
    52: '울산',
    53: '대구',
    54: '경북',
    55: '경남',
    61: '전남',
    62: '광주',
    63: '전북',
    64: '제주'
    # 다른 코드와 지역 이름도 추가해주세요
}

# '도착지(ARI)' 칼럼의 데이터를 지역 이름으로 변환
df['ARI'] = df['ARI'].map(ari_mapping)

# 변환된 '도착지(ARI)'별 데이터 수 계산
ari_counts = df['ARI'].value_counts()

# 그래프 그리기
plt.figure(figsize=(10, 6))
ari_counts.plot(kind='bar', color='orange')
plt.title('Number of Data by Arrival')
plt.xlabel('Arrival')
plt.ylabel('Data Count')
plt.xticks(rotation=45)
plt.tight_layout()

# 막대 위에 데이터 수 표시
for i, v in enumerate(ari_counts):
    plt.text(i, v + 50, str(v), ha='center', va='bottom', fontsize=8)

plt.show()
```
